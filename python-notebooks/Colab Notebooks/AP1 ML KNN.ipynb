```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
```

```python
path_training ="/content/drive/MyDrive/Machine Learn/TrainingData_2F_Original.txt"
path_test ="/content/drive/MyDrive/Machine Learn/TestingData_2F_Original.txt"

df = pd.read_csv(path_training, sep='\t')
df_test= pd.read_csv(path_test, sep='\t')
#display(df)
ID = df.pop('ID')
ID_test = df_test.pop('ID')
classes = df.pop('class')
classes_test = df_test.pop('class')
display(df[0:2])
#display(df.columns)
#display(df['total.sulfur.dioxide'][0:2])

total_rows = df.shape[0]
total_rows_test = df_test.shape[0]
print(total_rows)
```

```python
def euclidean_distance(row1, row2):
	distance = 0.0
	for i in range(len(row1)):		
		distance = distance + (row1[i] - row2[i])**2
		#print(distance, row1[i], row2[i])
	return sqrt(distance)
```

```python
k=5
print(k)
predictions = []
for t in range(total_rows_test):
  instance_test = df_test.iloc[t]
  #display(t)
  #display(instance_test)
  
  distances = []
  for xk in range(total_rows):
    instance_train = df.iloc[xk]
    #display(instance_train)
    
    d = euclidean_distance(instance_train, instance_test)      
    distances.append((d,xk))   

  distances = sorted(distances)
  #print('Distances',distances)

  nearest_neigs = []
  for kay in range(k):
    aux = distances[kay]
    #print(kay,'-est neig:',aux)
    nearest_neigs.append(aux)

  nearest_neigs_indexes = []
  #print(nearest_neigs)    
  for l in nearest_neigs:
    #print(l)
    for i, item in enumerate(l):
      #print(i,item)
      if i==1:        
       nearest_neigs_indexes.append(item)
  print('nearest_neigs_indexes',nearest_neigs_indexes)

  #print('classes', classes)
  #print('classes_test', classes_test)

  one = 1
  zero = 0
  trues = 0
  falses = 0 

  for kay in range(k):
    index = nearest_neigs_indexes[kay]
    if classes[index] == 1:      
      trues = trues + 1
    else:
      falses = falses + 1
    
    #print(kay,':',classes[index])
  if trues > falses:
   predictions.append(1)
  else: predictions.append(0)

print(predictions)
  
acuracies = sum(1 for x,y in zip(predictions,classes_test) if x == y) / len(predictions)
print('acuracy:', acuracies)
```
