```python
from google.colab import drive
drive.mount('/content/drive')
from datetime import datetime
import json
import matplotlib.pyplot as plt
import numpy as np

```

```python

accuracies = []
accuracies2 = []
accuracies3 = []

for i in range(34):
  title =  "/content/drive/MyDrive/network-traffic-classification-main/nn-reports/df0_2022-09-01_with_%d"%(i+2) + "_attributes.json"
  with open(title, "r") as f:
    rep = json.load(f)
  accuracies.append(rep["accuracy_test"])
  
for i in range(34):
  title =  "/content/drive/MyDrive/network-traffic-classification-main/nn-reports/df1_2022-09-01_with_%d"%(i+2) + "_attributes.json"
  with open(title, "r") as f:
    rep = json.load(f)
  accuracies2.append(rep["accuracy_test"])
  
for i in range(34):
  title =  "/content/drive/MyDrive/network-traffic-classification-main/nn-reports/df3_2022-09-01_with_%d"%(i+2) + "_attributes.json"
  with open(title, "r") as f:
    rep = json.load(f)
  accuracies3.append(rep["accuracy_test"])



```

```python

print(accuracies)
max_acc = max(accuracies2)
min_acc = min(accuracies3)
len_acc = len(accuracies)

xc = range(2,len_acc+2)
plt.figure(figsize=[8, 6])
plt.plot(xc, accuracies)
plt.plot(xc, accuracies2)
plt.plot(xc, accuracies3)

plt.ylabel('Acurácia de teste')
plt.xlabel('Número de atributos')
plt.title('Acurácia de teste por número de atributos')

y=np.arange(0.05,0.9,0.025)
z=np.arange(min_acc,max_acc,0.005)

plt.xticks(range(len(accuracies)+2))
plt.yticks(y)
plt.ylim(min_acc-0.0025, max_acc+0.0025)
for i in y:
  plt.axhline(i, color="black", linestyle=":")
for i in range(0,len_acc+2,2):
  plt.axvline(i, color="black", linestyle=":")

# Date for report file
now = datetime.now()
dt_string = now.strftime("%d-%m-%Y %H:%M:%S")
title = "/content/drive/MyDrive/network-traffic-classification-main/nn-reports/KneeCurve_%s"%dt_string + ".pdf"
plt.savefig(title)
```
